### 2.3 Interpolation of Missing Values
Here, I will use interpolation_ strategies to estimate the missing values from other training examples. For that I have several possible methods such as:
- __mean imputation__ replace missing value with mean value of feature column
- __median__ replaces with the median of feature column
- __rolling mean__ replaces with rolling mean of upper and lower values

Since this dataset deals with hourly data of energy generation, consumption and prices, which is luckily already sorted by the first column entry "time" in the dataset, it makes more sense to make the missing values dependant on the values within a certain _time range_ , using a __rolling mean__ of +/- x values. This is because it is from a context unlikely that the energy generation in 1 hour spikes down to the mean value and up again one hour later. I chose x = 3, in order to ensure that the values are primarily dependant on the close data.

```python
# Rolling mean implementation
df = df.fillna(df.rolling(3,min_periods=1).mean())
#df.isnull().sum()
```

### 2.4 Reorganizing the Features
For my further analyses on the dataset, I reorganized some of the features in it. Especially the __time__ is of interest here, since for this projcet i am planning to predict a time series of energy prices. Therefore in the following I decomposed the time feature in two ways: 
1. First, decomposing it into its initial year, months, day and hour values
2. Second, by converting it into a datetime object, setting it as the df-index

```python
# Approach 1: Split the time variable into hour, day, month and year values
def date_to_hour(date):
    """
    Function that decomposes the time feature and returns the hour feature.
    params: date
    returns: hour
    """
    hour = float(date.split(" ")[1].split(":")[0])
    return hour

def date_to_day(date):
    """
    Function that decomposes the time feature and returns the day feature.
    params: date
    returns: day
    """
    day = float(date.split(" ")[0].split("-")[2])
    return day

def date_to_month(date):
    """
    Function that decomposes the time feature and returns the month feature.
    params: date
    returns: month
    """
    month = float(date.split(" ")[0].split("-")[1])
    return month

def date_to_year(date):
    """
    Function that decomposes the time feature and returns the year feature.
    params: date
    returns: year
    """
    year = float(date.split(" ")[0].split("-")[0])
    return year

# add decomposed time column to dataset
df["hour"] = df.time.apply(date_to_hour)
df["day"] = df.time.apply(date_to_day)
df["month"] = df.time.apply(date_to_month)

# Approach 2: Convert the time to datetime object and set it as index
df['time'] = pd.to_datetime(df['time'], utc=True, infer_datetime_format=True)
df = df.set_index('time')

# Check if it worked
df.info()
```

### 3.2 Time Series Plots

In this sub-section I will present several time series plots about my dataset. Within these plots you will already see information and trends that I will use later on when creating my prediction model. 


To plot the time series smoothly, i obtained a convenient plotting function for plotting the feature columns against their time index created earlier: __plot_time_series__.

```python
# create a function that plots time series 
def plot_time_series(df=None, 
                     column=None,
                     column1=None,
                     series=pd.Series([]), 
                     label=None,
                     label1=None,
                     ylabel=None, 
                     title=None, 
                     start=0, 
                     end=None,
                     width=None,
                     facecolor=None):

    """
    This function is a convenience function to plot time series of my features
    from the energy dataset. It takes the dataframe and columns or series as data,
    a label, title, start- and end value as well as the line width and color 
    as parameters and plots the time series in a standardized way
    params: df 
    params: column=None,
    params: column1=None,
    params: series=pd.Series([]), 
    params: label=None,
    params: label1=None,
    params: ylabel=None, 
    params: title=None, 
    params: start=0, 
    params: end=None,
    params: width=None,
    params: facecolor=None):
    """
    sns.set()
    fig, ax = plt.subplots(figsize=(30, 12))
    ax.set_xlabel('Time', fontsize=16)
    ax.set_facecolor('white')
    plt.grid(color='grey', linestyle='-.', linewidth=0.5)
    
    if column:
        ax.plot(df[column][start:end], label=label, color=facecolor,linewidth=width)
        ax.set_ylabel(ylabel, fontsize=16)
    if column and column1:
        ax.plot(df[column1][start:end], label=label1, color='orange',linewidth=width)
    
    if series.any():
        ax.plot(series, label=label, color=facecolor)
        ax.set_ylabel(ylabel, fontsize=16)
        
    if label:
        ax.legend(fontsize=16)
        
    if title:
        ax.set_title(title, fontsize=24)

        
    ax.grid(True)
    return ax
```

Having created the convenience plot function, I will now do several analytic plots about the time series of the energy prices. Below you see both the plots of the __price_day_ahead__ as well as the __price_actual__ for the same time period.

I will split the dataset in a __80/10/10__ ratio into training-, validation-, and test set.

These datasets are defined as follows:
1. __Training Set:__ Train the model on the training data.
2. __Validation Set:__ After training the model, I check how well it performs on the valid set.
3. __Test Set:__ When I have a final model (i.e., the model that has performed well on both training as well as validation set), I will evaluate it on the test set in order to get the performance and errors on unseen data.

```python
# create test and training dataset as well as cross-validation set
train_end_idx = round(35039 * 0.8) 
valid_end_idx = round(35039 * 0.9) 
test_end_idx = 35064

df_regression_train = df_regression[:train_end_idx]
df_regression_valid = df_regression[train_end_idx : valid_end_idx]
df_regression_test = df_regression[valid_end_idx:]

print('Indices:',train_end_idx,
      valid_end_idx,
      test_end_idx)

print('\ndf_regression_train has ' + str(len(df_regression_train)) + ' rows.')
print('df_regression_test has ' + str(len(df_regression_test)) + ' rows.')
print('df_regression_valid has ' + str(len(df_regression_valid)) + ' rows.')
```

```python
# define data: response vector y
y_regression_train = df_regression_train['price_actual']
y_regression_valid = df_regression_valid['price_actual']
y_regression_test = df_regression_test['price_actual']

# define  data: feature matrix X (without response and response forecast (price_day_ahead))
X_regression_train = sm.add_constant(df_regression_train.iloc[:, 1:15])
X_regression_test = sm.add_constant(df_regression_test.iloc[:, 1:15])
X_regression_valid = sm.add_constant(df_regression_valid.iloc[:, 1:15])

# get names of feature columns
names = X_regression_train.columns.values

print(names,
      type(names))
```

```python
# Construct model estimate and summarize
model_rg = sm.OLS(y_regression_train,
                  X_regression_train)

estimate_rg = model_rg.fit()
print(estimate_rg.summary())
```

### 5.1 Statistical Test on the Dataset

Before I go deeper into creating the LSTM for this dataset, I will first conduct a test to clarify whether the time series in the dataset indeed contains a trend that can be analyzed with a time series forecast. The test I will use is the Dickey-Fuller Test.

#### Augmented Dickey-Fuller Test
The Augmented Dickey-Fuller Test, determines how strongly a time series is defined by a trend. The intuition behind the test is that if the series is characterised by a unit root process then the lagged level of the series $y_{t-1}$ will provide __no relevant information in predicting the change in $y_{t}$  besides the one obtained in the lagged changes $Delta(y_{t-k})$__ .

In this case the null hypothesis is not rejected. In contrast, when the process has no unit root, it is stationary and hence exhibits reversion to the mean - so the lagged level will provide relevant information in predicting the change of the series and the null of a unit root will be rejected.

- Null Hypothesis $H_{0}$: There is a unit root in the time series, i.e. the series is autocorrelated with (r = 1), a time dependent structure and thus, is not stationary.
- Alternate Hypothesis $H_{1}$: The time series has no unit root and is stationary.

More information about the test can be found here: https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test

```python
# create Dickey-Fuller dataset
y_Dickey = df_clean['price_actual']

# conduct Dickey-Fuller test
dickey_test = adfuller(y_Dickey, regression='c')

# print test results
print('Augmented Dickey-Fuller Statistic: {:.6f}\np-value: {:.6f}\n'
      .format(dickey_test[0], 
              dickey_test[1], 
              dickey_test[2]))

for key, value in dickey_test[4].items():print('Critical Value ({}): {:.6f}'.format(key, value))
```

#### Results Dickey-Fuller Test
The Augmented Dickey-Fuller Statistic of -9.074 is less than the critical value (-3.431) even for the 99% confidence interval. Thus, the null hypothesis $H_{0}$ can be rejected with a significance level 1%, meaning that there is not a root-unit in the time series and thus, that it is either stationary or could be made stationary with 1st order differencing (difference-stationary).

Since the Dickey-Fuller Test showed that the time series is stationary, the scaling method for the LSTM approach will be __normalization,__ which is sufficient for the further analyses.

### 5.2 Additional Feature Generation for the Time Series Forecast

I already decomposed the time feature of the energy dataset. To enable my model to learn the importance of the difference that can be caused by different hours of the day and days of the week, I will split the features further. That split will be conducted on the __df_clean__ dataset, which already holds the __hour__ , __day__ and __month__ feature

```python
# feature generation: See df_clean dataset: I already added the hour, day and month feature.
df_clean
```

__Working hours:__ Energy demand and therefore energy price is usually strongly dependent on whether businesses are open and people are working or not in a given hour. My initial approch to this was to align this feature with the classic German working hours. However, in Spain the 9-to-5 working day is not generally followed by all types of businesses, due to a lunch break in between ("Siesta").  The usual business and work hours are Monday-Saturday, from 9:00 - 14:00  and then again from 16:00 - 20:00 PM. 

The value of the __work_hour__ feature will be equal to 2 if the given hour is within the business hours, 1 if the given hour is within the "siesta" in between and equal to '0' for all other hours.

__Weekends:__ However, it is important to note that working hours on weekends might be different and that the work hours are not applicable at all on Sundays. Thus, I created another feature that will give the model the possibility to learn the influence of weekends on the energy price.

The value of the __weekend__ feature will be equal to 2 if the given day is a Sunday, 1 if it is a Saturday and equal to 0 for every other day.


```python
# create 'work_hour' feature -> add a feature column in dataset
for i in range(len(df_clean)):
    position = df_clean.index[i]
    
    hour = position.hour
    
    if ((hour > 8.9 and hour < 14) or (hour > 16 and hour < 20.1)):
        df_clean.loc[position, 'work_hour'] = 2
        
    elif (hour >= 14 and hour <= 16):
        df_clean.loc[position, 'work_hour'] = 1
        
    else:
        df_clean.loc[position, 'work_hour'] = 0

# create 'weekend' feature -> add a feature column in dataset
for i in range(len(df_clean)):
    position = df_clean.index[i]
    
    weekday = position.weekday()
    
    if (weekday == 6):
        df_clean.loc[position, 'weekday'] = 2 #Sunday
    elif (weekday == 5):
        df_clean.loc[position, 'weekday'] = 1 #Saturday
    else:
        df_clean.loc[position, 'weekday'] = 0 #weekday
```

The ourliers are then replaced by interpolation. In contrast to the earlier rolling mean interpolation, this time I will use linear interpolation.

```python
# remove outlier function: Replaces outliers with NaN values
for feature in considered_features:
    
    # calculate the mean of each feature in the clean dataset
    mean_feat = df_clean[feature].mean()
    
    # calculate standard deviation
    stdev_feat = df_clean[feature].std()
     
    # set lower and upper bond
    upper_limit = mean_feat + 5 * stdev_feat
    lower_limit = mean_feat - 5 * stdev_feat
    
    # replace outliers with NaN values
    df_clean.loc[df_clean[feature] > upper_limit, feature] = np.nan
    df_clean.loc[df_clean[feature] < lower_limit, feature] = np.nan
```

```python
# univariate data representation as input for my LSTM
def univariate_data(dataset, 
                    start_index, 
                    end_index, 
                    history_size, 
                    target_size):
    """
    This function creates the input data for my LSTM model. It gets a dataset,
    start- and end- index of the considered observations, the history size and
    a target size, and returns two arrays of data and labels.
    params: dataset
    params: start_index
    params: end_index
    params: history_size
    params: target_size
    returns: data, labels
    """
    data = []
    labels = []

    start_index = start_index + history_size
    if end_index is None:
        end_index = len(dataset) - target_size

    for i in range(start_index, end_index):
        indices = range(i - history_size, i)
        
        # Reshape data from (history_size,) to (history_size, 1)
        data.append(np.reshape(dataset[indices], (history_size, 1)))
        labels.append(dataset[i + target_size])
        
    return np.array(data), np.array(labels)

def plot_LSTM_rmse_and_loss(history_rmse_loss=None,
                            facecolor=None):
    """
    This function is a convenience function to plot the RMSE as well as the 
    loss function of the LSTM during training and validating.
    params: history_rmse_loss
    params: facecolor    
    """
    # set plot variables
    train_rmse = history.history['root_mean_squared_error']
    val_rmse = history.history['val_root_mean_squared_error']
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    # Visualize epochs vs. train and validation accuracies and losses
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20, 10))

    # design ax1 plot (RMSE)
    ax1.set_xlabel('Number of Epochs', fontsize = 14)
    ax1.set_ylabel('Training and Validation RMSE', fontsize = 14)
    ax1.plot(train_rmse, label='Training RMSE', linewidth = 2.0, color='#cb575f')
    ax1.plot(val_rmse, label='Validation RMSE', linewidth = 2.0, color = 'orange')
    ax1.set_facecolor('white')
    ax1.grid(color='grey', linestyle='-.', linewidth=0.7)
    ax1.set_title('Epochs vs. Training and Validation RMSE', fontsize = 16)
    
    # design ax2 plot (Loss)
    ax2.set_xlabel('Number of Epochs', fontsize = 14)
    ax2.set_ylabel('Training and Validation Loss', fontsize = 14)
    ax2.plot(train_loss, label='Training Loss', linewidth = 2.0, color='#cb575f')
    ax2.plot(val_loss, label='Validation Loss', linewidth = 2.0, color = 'orange')
    ax2.set_facecolor('white')
    ax2.grid(color='grey', linestyle='-.', linewidth=0.7)
    ax2.set_title('Epochs vs. Training and Validation Loss', fontsize = 16)

    plt.show()
    

def plot_learning_rate_schedule(history_learning_rate=None, 
                                max_loss=None, 
                                facecolor=None,
                                width = None):
    """
    This function is a convenience function to plot the learning rate 
    versus the training loss after training the LSTM.
    params: history_learning_rate
    params: max_loss
    params: facecolor
    params: width
    """
    fig, ax = plt.subplots(figsize=(30, 12))
    ax.set_xlabel('Learning Rate', fontsize = 30)
    ax.set_ylabel('Training Loss', fontsize = 30)
    plt.grid(color='grey', linestyle='-.', linewidth=0.7)
    
    ax.semilogx(history_learning_rate.history['lr'], 
                history_learning_rate.history['loss'], 
                color = facecolor,
                linewidth = width)
    plt.axis([1e-4, 1, 0, max_loss])
    ax.set_facecolor('white')

    plt.show()
```

#### Definition of Keras LSTM Model

Before defining the used LSTM Model in the code, I want to explain which model I will use and why. To forecast the time series of the actual energy price, I will conduct an hour-ahead forecast, meaning to forecast the energy price of the next hour based on a given set of past hours.

I will use the Sequential model from keras, which allows me to create a deep learning model layer-by-layer. It's limitations of e.g. not being able to re-use layers are not applicable for this problem. 

I used the keras __LSTM-Layer__, for which I set the unit number to 80 as well as the batch size to 32, which was given as a recommendation by several sources regarding LSTM tuning.

As input shape I give the model the input_shape_univ parameter. I also decided to flatten the input via a flatten-layer, which does not affect the batch size. In addition, I added a __dropout layer,__ which randomly sets input units to 0 with a frequency of 0.05 each step during training, which helps to prevent overfitting. Inputs not set to 0 are scaled up by 1/(1 - 0.05) so that the sum over all inputs is unchanged.

The default __activation function__ for this model is the tanh. I chose the relu activation function because it performed well in trial and error testing.

I chose to use keras __dense layer__ for this LSTM. As dimensionality of the output space of the dense layer (units) I chose 42, as well as the relu activation function. The number of layers in the neural network is 5.

The number of Epochs have been set to 50 as this turned out to be a convenient number where the model converged to a low loss during training.

#### Number of Layers

Although there is no generalistic rule for determining the optimal number of layers, I followed three common rules of thumb when trying different numbers of layers of the network:

1. The __complexity of the dataset,__ such as the number of features, the number of data points, etc: The energy dataset has rather many observations and features. This has to be taken into account.

2. The __data-generating process.__ For example, the prediction of oil prices compared to the prediction of GDP is a well-understood economy. The latter is much easier than the former. Thus, predicting oil prices might require more LSTM memory cells to predict, with the same accuracy, as compared to the GDP: From the earlier regression analyses as well as the baseline predictions you saw that the energy prices are (similar to oil prices) not a very well-understood economy. This has to be taken into account.

3. The __accuracy required for the use case.__ The number of memory cells will heavily depend on this. If the goal is to beat the state-of-the-art model, in general, one needs more LSTM cells: My goal is indeed to beat all baseline forecast shown earlier, especially the TSO's forecast. This has to be taken into account.

After evaluating the criteria and testing the performance by trial and error, I chose a total number of 5 layers which led to good results for my forecast.

```python
#define keras sequential LSTM Model
univariate_lstm = tf.keras.models.Sequential([
    LSTM(80, input_shape=input_shape_univ, return_sequences=True),
    Flatten(),
    Dense(42, activation='relu'),
    Dropout(0.05),
    Dense(1)
])

# define LSTM optimizer function
optimizer = tf.keras.optimizers.Adam(lr=1e-4, amsgrad=True)

# configure the model with losses and metrics
univariate_lstm.compile(loss=loss,
                        optimizer=optimizer,
                        metrics=metric)


model_checkpoint = tf.keras.callbacks.ModelCheckpoint('univariate_lstm.h5', save_best_only=True)

# define history
history_lr = univariate_lstm.fit(train_univ, epochs=50,
                                 validation_data=valid_univ,
                                 callbacks=[early_stopping, model_checkpoint])
```

#### Difference Univariate and Multivariate LSTM
The biggest difference between the univariate and multivariate LSTM is that there is more than one kind of observation for each time step, meaning that the model assumes multivariate time series data. There are two main models that are suitable to analyze multivariate time series data; Multiple Input Series (MIS) and Multiple Parallel Series (MPS). 

An MIS problem may have two or more parallel input time series and one output time series that is dependent on the input time series. An MPS problem, on the other hand, is the case where there are multiple parallel time series and a value must be predicted for each.

Here, I have a MIS problem since I only want to predict a single time series (price_actual) based on several input series. The MIS LSTM I will use gives me the ability to include more features into the forecast. For this, I will utilize the features I have created before in the __considered_features__ list by using the corresponding features from __df_LSTM.__

The procedure for model fitting and testing will be done in analogy to the procedure for the univariate LSTM.

The general Idea for the multivariate LSTM has been taken from a fellow contributor on github for a multivariate time series forecast. I then modified the implementation to align it with my previously implemented functions. Note that I have also adjusted the learning parameters in accordance with the previous univariate LSTM approach to ensure better comparability between the models. 

More information about multivariate LSTM can be found here: https://machinelearningmastery.com/multivariate-time-series-forecasting-lstms-keras/

The multivariate LSTM implementation can be viewed below. 

```python
# define multivariate input data analogue to univariate
def multivariate_data(dataset, 
                      target, 
                      start_index, 
                      end_index, 
                      history_size,
                      target_size, 
                      step, 
                      single_step=False):
    data = []
    labels = []

    start_index = start_index + history_size
    if end_index is None:
        end_index = len(dataset) - target_size
        
    for i in range(start_index, end_index):
        indices = range(i-history_size, i, step)
        data.append(dataset[indices])
        
        if single_step:
            labels.append(target[i + target_size])
        else:
            labels.append(target[i : i + target_size])

    return np.array(data), np.array(labels)

dataset = df_LSTM

# scaling analogue to univariate LSTM
scaler_mult = MinMaxScaler(feature_range=(0, 1))
scaler_mult.fit(dataset[:train_end_idx])
y = df_LSTM['price_actual'].values
scaler = MinMaxScaler(feature_range=(0, 1))
y_reshaped = y.reshape(-1, 1)
scaler.fit(y_reshaped[:train_end_idx])
scaled_price = scaler.transform(y_reshaped)
scaled_dataset = scaler_mult.transform(dataset)
scaled_dataset = np.concatenate((scaled_dataset, scaled_price), axis=1)

# Set number of time lags according to univariate LSTM
m_p_history = 24

m_f_target = 0

X_train_mult, y_train_mult = multivariate_data(scaled_dataset, scaled_dataset[:, -1],
                                               0, train_end_idx, 
                                               m_p_history,
                                               m_f_target,
                                               step=1, single_step=True)

X_val_mult, y_val_mult = multivariate_data(scaled_dataset, scaled_dataset[:, -1],
                                           train_end_idx, val_end_idx, 
                                           m_p_history,
                                           m_f_target,
                                           step=1, single_step=True)

X_test_mult, y_test_mult = multivariate_data(scaled_dataset, scaled_dataset[:, -1],
                                             val_end_idx, test_end_idx,
                                             m_p_history,
                                             m_f_target,
                                             step=1, single_step=True)

batch_size = 32
buffer_size = 1000

train_mult = tf.data.Dataset.from_tensor_slices((X_train_mult, y_train_mult))
train_mult = train_mult.cache().shuffle(buffer_size).batch(batch_size).prefetch(1)
val_mult = tf.data.Dataset.from_tensor_slices((X_val_mult, y_val_mult))
val_mult = val_mult.batch(batch_size).prefetch(1)

# Definition of common parameters
input_shape_mult = X_train_mult.shape[-2:]
loss = tf.keras.losses.MeanSquaredError()
metric = [tf.keras.metrics.RootMeanSquaredError()]
lr_schedule = tf.keras.callbacks.LearningRateScheduler(
              lambda epoch: 1e-4 * 10**(epoch / 10))
early_stopping = tf.keras.callbacks.EarlyStopping(patience=10)

y_test_mult_reshaped = y_test_mult.reshape(-1, 1)
y_test_mult_inv = scaler.inverse_transform(y_test_mult_reshaped)



# define multivariate LSTM with adjusted parameters
tf.keras.backend.clear_session()
multivariate_lstm = tf.keras.models.Sequential([LSTM(80, 
                                                     input_shape=input_shape_mult, 
                                                     return_sequences=True),
                                                     Flatten(),
                                                     Dense(42, activation='relu'),
                                                     Dropout(0.05),Dense(1)
])

optimizer = tf.keras.optimizers.Adam(lr=1e-4, amsgrad=True)
multivariate_lstm.compile(loss=loss,
                          optimizer=optimizer,
                          metrics=metric)


lr = multivariate_lstm.fit(train_mult, epochs=50,
                                   validation_data=val_mult,
                                   callbacks=[lr_schedule])
```

```python
# plot the training loss vs. the learning rate
ax = plot_learning_rate_schedule(history_learning_rate = lr,
                                 max_loss=0.01,
                                 facecolor='#cb575f',
                                 width=2.0)

plt.show()
```
